```python
#question 3

from qiskit import QuantumCircuit
from qiskit import QuantumRegister
from qiskit import ClassicalRegister
from qiskit.quantum_info import Statevector
from qiskit import transpile
from qiskit_aer import AerSimulator
from qiskit.visualization import plot_histogram
def superdensecoding(msg): 
    qr= QuantumRegister(2, name='q')
    cb= ClassicalRegister(2, name='b')
    circ= QuantumCircuit(qr,cb)
    circ.h(0)
    circ.cx(0,1)
    if msg==(0,0):
        pass
    elif msg==(0,1):
        circ.z(0)
    elif msg==(1,0):
        circ.x(0)
    else:
        circ.x(0)
        circ.z(0)
    circ.cx(0,1)
    circ.h(0)
    circ.measure(range(2),range(2))
    return circ
def word_to_binary(word):
    binary = ''.join(format(ord(c), '08b') for c in word)
    return binary
def binary_to_word(binary_str):
    padded = binary_str.ljust((len(binary_str) + 7) // 8 * 8, '0')
    return ''.join(chr(int(padded[i:i+8], 2)) for i in range(0, len(padded), 8))
msg=input("Enter message to be delivered: ")
str=''
num=word_to_binary(msg)
for i in range(0,len(num),2):
    bit_pair=[tuple(map(int, num[i:i+2]))]
    circ=superdensecoding(bit_pair)
    backend= AerSimulator()
    compiled_circ = transpile(circ, backend)
    job = backend.run(compiled_circ, shots=1024)
    result = job.result()
    counts = result.get_counts()
    most_frequent = max(counts, key=counts.get)
    str+=most_frequent
out=binary_to_word(str)
print(out)

```
